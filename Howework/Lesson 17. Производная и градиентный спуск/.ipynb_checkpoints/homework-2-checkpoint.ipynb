### Задание 1

Посчитайте 1-ую и 2-ую производные функции (на бумажке):

$f(x) = sin^2(2x+1)$


### Решение

$f'(x) = 2*sin(2x+1) * (sin(2x + 1))' = 2*sin(2x+1) * cos(2x + 1) * (2x + 1)' = 4 * sin(2x+1)*cos(2x + 1) $

$f''(x) = 4 * (sin(2x + 1)' * cos(2x + 1) + sin(2x + 1) * cos(2x +1)') $

$f''(x) = 4 * (cos(2x + 1) * 2 * cos(2x + 1) + sin(2x + 1) * (-sin(2x + 1) * 2) $

$f''(x) = 4 * (2 * cos^2(2x + 1) - 2 * sin^2(2x + 1)) $

$f''(x) = 8 * cos^2(2x + 1) - 8 * sin^2(2x +1) $

```python
x = np.array([i for i in range(-100, 100)])
y = np.array([math.sin(math.radians(2 * x_i + 1))**2 for x_i in x])
plt.plot(x, y)
```

```python
#первая производная
list_der_sin = []
for i in range(len(x)-1):
    list_der_sin.append((y[i+1]-y[i])/(x[i+1]-x[i]))
list_der_sin.append(list_der_sin[-1])
plt.plot(x, list_der_sin)
```

```python
def f(x):
    return math.sin(math.radians(2 * x + 1))**2
```

```python
#Первая производная (другой способ)
x = np.array([i for i in range(-100, 100)])
y = np.array([derivative(f, i, dx=1e-6) for i in x])
plt.plot(x, y)
```

```python
#вторая производная
list_second_der = []
for i in range(len(x)-1):
    list_second_der.append((y[i+1]-y[i])/(x[i+1]-x[i]))
list_second_der.append(list_second_der[-1])
plt.plot(x, list_second_der)
```

### Задание 2: 
Посчитайте частные производные функции в точкe
(на бумажке или в LaTeX и на numpy):

$f(x,y)=2x^2y^3 + 1/x + y^2x + 7$ в точке $(1,2)$

### Решение
$f'x(x,y)= 4xy^3 - 1/x^2 + y^2$


$f'y(x,y)= 6x^2y^2 + 2xy$

$f'x(1,2)= 4 * 1 * 8  - 1 + 4 = 35$

$f'y(1,2)= 6 * 1 * 4 + 1 + 2 * 1 * 2 = 28$

```python
def func(x, c0, c1, c2):
    return c0 * x[0]**2 * x[1]**3 + c1 * 1/x[0] + c2 * x[0] * x[1]**2 + 7
```

```python
x = np.array([1., 2.])
c0, c1, c2 = (2., 1., 1.)
eps = np.sqrt(np.finfo(float).eps)
approx_fprime(x, func, [eps, eps], c0, c1, c2)
```
