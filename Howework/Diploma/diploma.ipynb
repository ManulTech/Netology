### С использованием t-test статистики проверить взаимосвязи между несколькими переменными

Для применения t-тестов необходимо, чтобы распределение величины было нормальным или близким к нормальному. Поэтому тесты провожу для number_project, так как оно похоже на нормальное распределение
#### Гипотеза №1: проекты satisfied сотрудников не отличается от проектов not_satisfied сотрудников

Разбиваю satisfaction_level на две группы:
    1. >= 0,7 - хороший уровень удовлетворённости 
    2. < 0,7 - плохой уровень удовлетворённости

```python
def f(param):
    if param >= 0.7:
        return 1
    else:
        return 0
```

```python
df['satisfied'] = df['satisfaction_level'].apply(f)
```

```python
import scipy.stats as sts
import researchpy as rp
```

```python
satisfied=df[df['satisfied']==1]['number_project']
not_satisfied=df[df['satisfied']==0]['number_project']
```

```python
sts.ttest_ind(satisfied, not_satisfied)
```

```python
descriptives, results = rp.ttest(satisfied, not_satisfied)

descriptives
```

```python
results
```

#### Вывод: так как p-value < 0,05 различие между проектами (0.1362) статистически значимо. Вероятно конкретный проект влияет на уровень удовлетворённости сотрудников

#### Гипотеза №2: average_montly_hours  не зависит от количества проектов 

Для этого разобью average_montly_hours на две группы:

1: те, кто работает больше среднего >200+ часов

2: те, кто работает <= 200 часов

```python
def f(param):
    if param > 200:
        return 'above average'
    else:
        return 'below average'
```

```python
df['average_montly_hours_group'] = df['average_montly_hours'].apply(f)
df.head()
```

```python
above_average = df[df['average_montly_hours_group'] == 'above average']['number_project']
below_average = df[df['average_montly_hours_group'] == 'below average']['number_project']
```

```python
descriptives, results = rp.ttest(above_average, below_average)

descriptives
```

```python
results
```

#### Вывод: так как p-value < 0,05, можно сделать вывод, что выборки отличаются значимо. Коэффициент Пирсона 0.3379 показывает слабую корреляцию между количеством проектов и временем на работе. Так как он положительный, то чем больше проектов, тем больше времени сотрудник проводит на работе

### Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

Смотрим, что там с количеством времени, проведённым в компании и уходом из неё

```python
(df
    .groupby(['time_spend_company', 'left'])
    .agg({'satisfaction_level': 'count'})
    .sort_values(['satisfaction_level'], ascending=[False])
)
```

#### Вывод: получается, что сотрудникам нужно 3 года, чтобы понять, что пора валить из конторы. В зоне риска сотрудники, которые работают от 3х до 5 лет

Смотрим, что там с удовлетворённостью и и уходом из компании

```python
(df
    .groupby(['satisfaction_level', 'left'])
    .agg({'satisfaction_level': 'count'})
    .sort_values(['left'], ascending=[False])
)
```

```python
(df
    .groupby(['satisfied', 'left'])
    .agg({'satisfaction_level': ['count', 'mean']})
    .sort_values(['left'], ascending=[False])
)
```

#### Вывод: в большей степени уходят те, кто не удовлетворён

Посмотрим, что там с повышением, временем в компании и проверокой и увольнением

```python
(df
    .groupby(['time_spend_company', 'promotion_last_5years', 'left'])
    .agg({'satisfaction_level': ['mean', 'count']})
    .sort_values(['left'], ascending=[False])
)
```

####  Выводы: 
 - повышение не влияет на решение покинуть компанию
 - повышение не влияет на общий уровень удовлетворённости
 - больше всех уходят сотрудники, проработавшие 3-4 года. Уровень удовлетворённости 50% и меньше. Стоит обратить внимание на эту группу
 - сотрудники, которые работали 5+ лет лояльны к компании (0,6+), но вероятнее всего есть другие внешние факторы, которые заставляют их покинуть компанию
