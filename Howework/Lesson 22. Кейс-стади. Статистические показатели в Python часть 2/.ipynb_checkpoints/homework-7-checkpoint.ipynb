### Между возрастными группами

```python
filt_data['age'].unique
```

```python
filt_data['age_years'] = filt_data['age'] / 365
```

```python
filt_data
```

```python
def age_group(param):
    if param < 21:
        return 1
    elif param < 35:
        return 2
    elif param < 60:
        return 3
    else:
        return 4
```

```python
filt_data['age_years_group'] = filt_data['age_years'].apply(age_group) 
```

```python
filt_data
```

```python
filt_data['age_years_group'].unique()
```

```python
age_group_3 = filt_data[filt_data['age_years_group']==3].cardio
age_group_4 = filt_data[filt_data['age_years_group']==4].cardio
descriptives, results = rp.ttest(age_group_3, age_group_4)
print(descriptives)
print('_____________________________')
print(results)
```

### Random Forest

```python
x, y = filt_data.drop(['cardio', 'age_years', 'age_years_group'], axis=1), filt_data.cardio
```

```python
x.head()
```

```python
y.head()
```

### Нормализация данных

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))

np_scaled = min_max_scaler.fit_transform(x)
df_norm = pd.DataFrame(np_scaled)
df_norm.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_norm, y, test_size = 0.2, random_state = 50)
```

```python
from sklearn.ensemble import RandomForestClassifier

# создаем модель деревья решений
# выбираем 100 деревьев в качестве параметра
model=RandomForestClassifier(n_estimators=100)

# обучаем модель
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
model_pred = model.predict(x_test)

print(classification_report(y_test, model_pred))
```

### Вывод:

В 70% случаев модель предскажет правильно (диагностирует заболевание). Но в 33% (100 - 67) случаях отнесёт здорового пациента к больным. Правильно?
