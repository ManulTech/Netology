```python
import pandas as pd
```

```python
ratings = pd.read_csv(r'C:\Users\aekom\Lessons\14 Продвинутый pandas 2\Python_13_join\ml-latest-small\ratings.csv')
```

```python
ratings_count = ratings.groupby('userId').agg({'timestamp': ['min', 'max', 'count']})
```

## Решение 2

```python
ratings_min = ratings.groupby('userId').min()[['timestamp']]
ratings_max = ratings.groupby('userId').max()[['timestamp']]
ratings_100 = ratings.groupby('userId').count()
ratings_100 = ratings_100[ratings_100.movieId > 100]
ratings_100
```

```python
ratings_min.rename(columns={'timestamp': 'min'}, inplace=True)
ratings_max.rename(columns={'timestamp': 'max'}, inplace=True)
ratings_min.head()
```

```python
ratings_100 = ratings_100.merge(ratings_min, on='userId', how='left').merge(ratings_max, on='userId', how='left').reset_index()
ratings_100.head()
```

```python
ratings_100['Time difference'] = ratings_100['max'] - ratings_100['min']
```

```python
ratings_100
```

```python
ratings_100_2 = (
    ratings_100_2
    .join(ratings_min_2, how='left', on='userId', lsuffix='_left')
    .join(ratings_max_2, how='left', on='userId', lsuffix='_left')
)
ratings_100_2
```

```python
ratings_100_2['time_diff'] = ratings_100_2['max'] - ratings_100_2['min']
ratings_100_2
```
